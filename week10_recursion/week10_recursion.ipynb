## Part 2 — Core Concepts (10–15 min)

### What Is Recursion?

Recursion is when a function calls **itself** to solve a smaller version of the same problem. Every recursive function needs two things:

| Component | Purpose |
|-----------|---------|
| **Base Case** | The condition that **stops** the recursion. Without it, the function runs forever. |
| **Recursive Case** | The part where the function **calls itself** with a smaller or simpler input. |

Think of it like cracking a combination lock: you try one digit, then hand the remaining digits to a helper — who does the same thing until there are no digits left.

### Example: Factorial

`factorial(n)` = n × (n-1) × (n-2) × ... × 1

- **Base case:** `factorial(1) = 1` (we know the answer without more work)
- **Recursive case:** `factorial(n) = n * factorial(n - 1)`

````python
def factorial(n):
    # Base case
    if n == 1:
        return 1
    # Recursive case
    return n * factorial(n - 1)

print(factorial(5))   # 120
print(factorial(3))   # 6
````

### How the Call Stack Works

When `factorial(4)` runs, Python builds a **call stack**:

```
factorial(4)  -->  waits for factorial(3)
  factorial(3)  -->  waits for factorial(2)
    factorial(2)  -->  waits for factorial(1)
      factorial(1)  -->  returns 1       (base case!)
    factorial(2)  -->  returns 2 * 1 = 2
  factorial(3)  -->  returns 3 * 2 = 6
factorial(4)  -->  returns 4 * 6 = 24
```

Each call **pauses**, waiting for the inner call to return its result. Then results **bubble back up** the chain.

### Example: Sum of a List

Here is a recursive way to sum all elements of a list:

````python
def recursive_sum(lst):
    # Base case: empty list has sum 0
    if len(lst) == 0:
        return 0
    # Recursive case: first element + sum of the rest
    return lst[0] + recursive_sum(lst[1:])

print(recursive_sum([10, 20, 30, 40]))  # 100
````

### When to Use Recursion?

Recursion is especially useful when the problem has a **naturally recursive structure**:

- Traversing directories and file systems (common in forensics)
- Parsing nested data (JSON, XML, network packets)
- Generating all combinations (brute-force password cracking)
- Tree/graph exploration (network mapping)

If the problem can be broken into **identical smaller sub-problems**, recursion is usually a clean fit.

### Exercise 1: Recursive String Reversal (Beginner)

In cybersecurity, you sometimes need to reverse-engineer encoded strings. A simple starting point is reversing a string.

**Your task:** Write a function called `reverse_string(s)` that takes a string `s` and returns it reversed using **recursion** (do NOT use slicing like `s[::-1]` or the built-in `reversed()`).

**Think about:**
- What is the simplest string you can reverse without doing any work? (That is your base case.)
- How can you break a string into a smaller piece and combine the results?

**Examples:**
```
reverse_string("abc")     -->  "cba"
reverse_string("hello")   -->  "olleh"
reverse_string("a")       -->  "a"
reverse_string("")        -->  ""
```

````python
def reverse_string(s):
    if len(s) <= 1:
      return s
    return reverse_string(s[1:]) + s[0]


# Test your function
print(reverse_string("abc"))      # Expected: cba
print(reverse_string("hello"))    # Expected: olleh
print(reverse_string("a"))        # Expected: a
print(reverse_string(""))         # Expected: (empty string)
````

### Exercise 2: Recursive Power Function (Intermediate)

Exponentiation is at the heart of many cryptographic algorithms (RSA, Diffie-Hellman). In this exercise you will compute `base^exp` recursively.

**Your task:** Write a function called `power(base, exp)` that computes `base` raised to the power `exp` using recursion. Assume `exp` is always a non-negative integer.

**Rules:**
- Do NOT use the `**` operator or `pow()`.
- Use only multiplication and recursive calls.

**Think about:**
- What is anything raised to the power 0?
- How can `base^exp` be expressed in terms of `base^(exp-1)`?

**Examples:**
```
power(2, 10)  -->  1024
power(3, 4)   -->  81
power(5, 0)   -->  1
power(7, 1)   -->  7
```

````python
def power(base, exp):
  if (exp == 0):
    return 1
  return base * power(base, exp - 1)


# Test your function
print(power(2, 10))   # Expected: 1024
print(power(3, 4))    # Expected: 81
print(power(5, 0))    # Expected: 1
print(power(7, 1))    # Expected: 7
````

### Exercise 3: Recursive Password Generator (Advanced)

Penetration testers sometimes need to generate all possible passwords of a given length from a set of characters. This is a classic recursion problem.

**Your task:** Write a function called `generate_passwords(chars, length)` that takes:
- `chars` — a string of allowed characters (e.g., `"abc"`)
- `length` — the desired password length

The function should **return a list** of all possible passwords of exactly that length.

**Think about:**
- What is the set of passwords of length 0? (Hint: there is exactly one — the empty string.)
- If you already have all passwords of length `(length - 1)`, how do you build passwords of length `length`?

**Examples:**
```
generate_passwords("ab", 1)   -->  ["a", "b"]
generate_passwords("ab", 2)   -->  ["aa", "ab", "ba", "bb"]
generate_passwords("abc", 1)  -->  ["a", "b", "c"]
```

**Warning:** Do not test with `length > 4` and a large character set — the output grows exponentially!

````python
def generate_passwords(chars, length):
    if length == 0:
        return [""]
    result = []
    for char in chars:
        for password in generate_passwords(chars, length - 1):
            result.append(char + password)
    return result


# Test your function
print(generate_passwords("ab", 1))    # Expected: ['a', 'b']
print(generate_passwords("ab", 2))    # Expected: ['aa', 'ab', 'ba', 'bb']
print(generate_passwords("abc", 1))
print(generate_passwords("bersu",5))  # Expected: ['a', 'b', 'c']
# Uncomment to test (expect 8 results):
# print(generate_passwords("ab", 3))
````

### Challenge: Recursive Directory Scanner

Security analysts often need to scan nested directory structures to find suspicious files. Below is a **simulated** file system represented as a Python dictionary. Directories map to dictionaries; files map to their size in bytes.

```python
file_system = {
    "home": {
        "user": {
            "documents": {
                "report.pdf": 52000,
                "notes.txt": 1200
            },
            "downloads": {
                "malware.exe": 98000,
                "image.png": 34000,
                "tools": {
                    "scanner.py": 4500,
                    "keylogger.exe": 67000
                }
            }
        }
    }

```

**Your task:** Write a recursive function called `find_files(fs, extension)` that searches through the entire simulated file system and returns a list of all file **names** (just the name, not the full path) that end with the given extension.

**Example:**
```
find_files(file_system, ".exe")  -->  ["malware.exe", "keylogger.exe"]
find_files(file_system, ".txt")  -->  ["notes.txt"]
```

**Think about:**
- How do you tell the difference between a file and a directory in this structure?
- What do you do when you encounter a directory? A file?

````python
file_system = {
    "home": {
        "user": {
            "documents": {
                "report.pdf": 52000,
                "notes.txt": 1200
            },
            "downloads": {
                "malware.exe": 98000,
                "image.png": 34000,
                "tools": {
                    "scanner.py": 4500,
                    "keylogger.exe": 67000
                }
            }
        }
    }
}

def find_files(fs, extension):
    found = []
    for key, value in fs.items():
        if isinstance(value, dict):
            found.extend(find_files(value, extension))
        elif key.endswith(extension):
            found.append(key)
    return found



# Test your function
print(find_files(file_system, ".exe"))   # Expected: ['malware.exe', 'keylogger.exe']
print(find_files(file_system, ".txt"))   # Expected: ['notes.txt']
print(find_files(file_system, ".py"))    # Expected: ['scanner.py']
````
